```python
import pandas as pd
from pandas.io.json import json_normalize

import seaborn as sns
sns.set_style('white')
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams["axes.formatter.useoffset"] = False
```

```python
# Load json file to dataframe

def load_fin(filename, year):
    df = pd.read_json(filename)
    df = df.ix[:, 1:72]
    df = df.transpose()
    df.reset_index(level=0, inplace=True)
    df.rename(columns = {'index':'organization'}, inplace=True)
    df = pd.melt(df, id_vars=['organization'])
    df.rename(columns = {'variable':'purpose'}, inplace=True)

    def to_list_dicts(x):
        x = str(x)
        if (x != 'nan') and (x is not None):
            return eval(x)
        else:
            return {}

    df['value'] = pd.DataFrame(df['value'].apply(to_list_dicts))

    data = []
    for index, row in df.iterrows():
        if len(row['value']) > 0:
            for i in row['value']:
                if (i['item'] is not None) or (i['sum'] > 0):
                    z = {'org': row['organization'], 'purpose': row['purpose'], 'year': year}
                    z.update(i)
                    data.append(z)

    return pd.DataFrame(data)

df14 = load_fin('data/finfin14.json', '2014')
df15 = load_fin('data/finfin15.json', '2015')
df16 = load_fin('data/finfin16.json', '2016')

df = pd.concat([df14, df15, df16], axis=0)
print df.shape
df.head()
```

```python
# Print costs 
def print_sum(df, n_rows):
    n = 1
    for _, row in df.iterrows():
        if n <= n_rows:
            print u"{}. {}\n{:,.2f} руб.\n{}\n{}г.\n".format(n, row['item'], row['sum'], row['org'], row['year'])
            n += 1
```

```python
# Конкретные расходы - самые дорогие
print_sum(df.sort_values(by='sum', ascending=False), 10)
```

```python
# Конкретные расходы - самые дешевые
print_sum(df.sort_values(by='sum'), 10)
```

```python
# Кто тратит больше всего
df_money = df[['org', 'sum']].groupby(['org']).agg(['sum', 'count'])
df_money.reset_index(level=0, inplace=True)
df_money.columns = df_money.columns.droplevel()
df_money.columns = ['org', 'sum', 'count']
df_money = df_money.sort_values(by='sum', ascending=False)
df_money['check'] = df_money['sum']/df_money['count']
n = 1
for _, row in df_money.iterrows():
    if n <= 10:
        print u"{}. {}\n{:,.2f} руб. / {} операций\n".format(n, row['org'], row['sum'], row['count'])
        n += 1
```

```python
# Средний чек - максимальный
n = 1
for _, row in df_money.sort_values(by='check', ascending=False).iterrows():
    if n <= 10:
        print u"{}. {}\nСредний чек {:,.2f} руб. / {} операций\n".format(n, row['org'], row['check'], row['count'])
        n += 1
```

```python
# Средний чек - минимальный
n = 1
for _, row in df_money.sort_values(by='check').iterrows():
    if n <= 10:
        print u"{}. {}\nСредний чек {:,.2f} руб. / {} операций\n".format(n, row['org'], row['check'], row['count'])
        n += 1
```

```python
# У кого больше всего операций
n = 1
for _, row in df_money.sort_values(by='count', ascending=False).iterrows():
    if n <= 10:
        print u"{}. {}\n\t{} операций".format(n, row['org'], row['count'])
        n += 1
```

```python
# Наиболее частые статьи расходов
df['item'] = df.item.str.lower()
df.item.value_counts().head(11)
```

```python
df_year = df[['sum', 'year', 'purpose']].groupby(['year', 'purpose']).agg(['sum', 'count'])
df_year.reset_index(level=0, inplace=True)
df_year.columns = df_year.columns.droplevel()
df_year.reset_index(level=0, inplace=True)
df_year.columns = ['purpose', 'year', 'sum', 'operations']
df_year['check'] = df_year['sum']/df_year['operations']

df_mln = df_year.copy()
df_mln['sum'] = df_mln['sum']/1000000
df_mln['check'] = df_mln['check']/1000000

df_mln
```

```python
# Динамика по годам
# Общая сумма расходов

plt.figure()
sns.factorplot(y="sum", x="year",  
                   data=df_mln[df_mln['sum'].notnull()], ci=None, size=5, aspect=1.3)
plt.ylabel(u'Расходы, млн. руб.')
plt.xlabel('')
plt.ylim(ymin=-100)
plt.savefig('plots/costs_year.png')
sns.plt.show()
```

```python
# Динамика по годам
# Общая сумма расходов по статьям
plt.figure()
g = sns.factorplot(y="sum", x="year", hue="purpose", 
                   data=df_mln[df_mln['sum'].notnull()], ci=None, size=5, aspect=1.3)
plt.ticklabel_format(style='plain', axis='y')
plt.ylim(ymin=-9000)
plt.ylabel(u'Расходы, млн. руб.')
plt.xlabel('')
plt.savefig('plots/costs_year_purpose.png')
plt.show()
```

```python
# Динамика по годам
# Check
plt.figure()
sns.factorplot(y="check", x="year",  
                   data=df_mln[df_mln['sum'].notnull()], ci=None, size=5, aspect=1.3)
plt.ylabel(u'Средний чек, млн. руб.')
plt.xlabel('')
plt.savefig('plots/check_year.png')
sns.plt.show()
```

```python
# Динамика по годам
# Operations
plt.figure()
sns.factorplot(y="operations", x="year",  
                   data=df_mln[df_mln['sum'].notnull()], ci=None, size=5, aspect=1.3)
plt.ylabel(u'Количество операций в отчете')
plt.xlabel('')
plt.savefig('plots/operations_year.png')
sns.plt.show()
```
